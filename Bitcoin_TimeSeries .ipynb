### Resample

```python
plt.figure(figsize=(12,8))
bitcoin['2021']['Close'].plot()
bitcoin['2021']['Close'].resample('M').mean().plot(label='Moyenne par mois',lw=3,ls=':',alpha=0.8)
bitcoin['2021']['Close'].resample('W').mean().plot(label='Moyenne par semaine',lw=2,ls='--',alpha=0.8)
plt.legend()
plt.show()
```

# Aggregate

```python
Ag=bitcoin['2021']['Close'].resample('W').agg(['mean','std','min','max'])
Ag
```

```python
plt.figure(figsize=(12,8))
Ag['mean']['2021'].plot(label='Moyenne par semaine')
plt.fill_between(Ag.index,Ag['min'],Ag['max'],alpha=0.2,label='min-max par semaine')
plt.legend()
plt.show()
```

# Moving Average

Au lieu de faire la moyenne de toutes les valeurs,in fait la moyenne sur une fenetre de valeurs

```python
plt.figure(figsize=(12,8))
bitcoin['2021']['Close'].plot()
bitcoin['2021']['Close'].rolling(window=7).mean().plot(label='Moving average',lw=3,ls='--')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))
bitcoin['2021-09']['Close'].plot()
bitcoin['2021-09']['Close'].rolling(window=7).mean().plot(label='Moving average',lw=3,ls='--')
bitcoin['2021-09']['Close'].rolling(window=7,center=True).mean().plot(label='Center Moving average',lw=3,ls=':')
plt.legend()
plt.show()
```

Le moving averrage c'est pas le bon model pour cette manipulation

$ Sol = Moyenne mobile Exponentielle $   
$ ewm(alpha=0.5).mean $

```python
plt.figure(figsize=(12,8))
bitcoin['2021-09']['Close'].plot()
bitcoin['2021-09']['Close'].rolling(window=7).mean().plot(label='Moving average',lw=3,ls='--')
bitcoin['2021-09']['Close'].rolling(window=7,center=True).mean().plot(label='Center Moving average',lw=3,ls=':')
bitcoin['2021-09']['Close'].ewm(alpha=0.6).mean().plot(label=' EWM ',lw=3,ls=':',c='yellow')
plt.legend()
plt.show()
```

# Turtle Strategy

1/utiliser $rolling()$ pour calculer :  
       * $max$ sur les 28 derniers jours   
       * $min$ sur les 28 derniers jours   
2/Boolean indexing:  
        *si $'Close' > max28 $ alors $BUY=1$  
        *si $'Close' < min28 $ alors $SELL=-1$
         


```python
data=bitcoin.copy()
data['BUY']=np.zeros(len(data))
data['SELL']=np.zeros(len(data))
```

```python
data['MaxRol']=data['Close'].shift(1).rolling(window=28).max()
data['MinRol']=data['Close'].shift(1).rolling(window=28).min()
data.loc[data['MaxRol'] < data['Close'],'BUY']=1
data.loc[data['MinRol'] > data['Close'],'SELL']=-1

```

```python
start='2021'
end='2021'
fig,ax=plt.subplots(2,figsize=(12,8),sharex=True)
ax[0].plot(data['Close'][start:end])
ax[0].plot(data['MinRol'][start:end])
ax[0].plot(data['MaxRol'][start:end])
ax[0].legend(['close','min','max'])
ax[1].plot(data['BUY'][start:end],c='g')
ax[1].plot(data['SELL'][start:end],c='r')
ax[1].legend(['BUY','SELL'])

```
